The purpose of this project's analysis is to identify trends in corn production, including corn production, land area used, and productivity measured annually. # Preprocessing

## Drop Index

```python
Corn_2020=Corn_2020.drop(index=[0,1,2,41])
Corn_2021=Corn_2021.drop(index=[0,1,2,41])
Corn_2022=Corn_2022.drop(index=[0,1,2,41])
Corn_2023=Corn_2023.drop(index=[0,1,2,41])
Corn_2024=Corn_2024.drop(index=[0,1,2,41])
```

## Add Year Column

```python
Corn_2020['Year']='2020'
Corn_2021['Year']='2021'
Corn_2022['Year']='2022'
Corn_2023['Year']='2023'
Corn_2024['Year']='2024'
```

## Combine Dataset

```python
Corn_analysis_df=pd.concat([
    Corn_2020,
    Corn_2021,
    Corn_2022,
    Corn_2023,
    Corn_2024
], ignore_index=True)
```

## Rename Column

```python
Corn_analysis_df=Corn_analysis_df.rename(columns={
    "38 Provinsi":'Province',
    'Unnamed: 1':'Harvest Area (ha)',
    'Unnamed: 2':'Productivity (ku/ha)',
    'Unnamed: 3':'Production (Ton)'
})
```

## Handling Data

```python
Corn_analysis_df['Harvest Area (ha)']=Corn_analysis_df['Harvest Area (ha)'].replace('-',0)
Corn_analysis_df['Productivity (ku/ha)']=Corn_analysis_df['Productivity (ku/ha)'].replace('-',0)
Corn_analysis_df['Production (Ton)']=Corn_analysis_df['Production (Ton)'].replace('-', 0)
```

```python
Corn_analysis_df['Harvest Area (ha)']=Corn_analysis_df['Harvest Area (ha)'].astype(float)
Corn_analysis_df['Productivity (ku/ha)']=Corn_analysis_df['Productivity (ku/ha)'].astype(float)
Corn_analysis_df['Production (Ton)']=Corn_analysis_df['Production (Ton)'].astype(float)
```

# EDA

## How has corn production trended in Indonesia from year to year?

```python
Annual_Corn=Corn_analysis_df.groupby(by='Year').agg({
    "Production (Ton)":"sum"
})
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(
    data=Annual_Corn,
    x=Annual_Corn.index,
    y='Production (Ton)',
    marker='o'
)
plt.title('Corn Production Trend In Indonesia Each Year')
plt.xlabel('Year')
plt.ylabel('Production (Ton)')
plt.grid(True)
plt.show()
```

## Insight
Based on the visualization above, it can be concluded that corn production in Indonesia fluctuated during the 2020 - 2024 period. Corn production showed an upward trend from 2020 - 2022. However, during the 2022 - 2023 period, there was a decline in production, before increasing again in the following year.

## What is the average corn harvest area from year to year in Indonesia?

```python
Mean_Harvest_Area=Corn_analysis_df.groupby(by='Year').agg({
    'Harvest Area (ha)':'mean'
})
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(
    data=Mean_Harvest_Area,
    x=Mean_Harvest_Area.index,
    y='Harvest Area (ha)',
    marker='o'
)
plt.title('Average Corn Harvest Area in Indonesia Each Year')
plt.xlabel('Year')
plt.ylabel('Harvest Area (ha)')
plt.grid(True)
plt.show()
```

### Insight

Based on the visualization of corn land area per year, it can be seen that the corn land area fluctuates. In the 20212022 period, there was an increase in corn land area, indicating that corn cultivation was quite intensive during that time frame. However, in the following year, the corn land area decreased.

## How has corn production productivity in Indonesia changed over time?

```python
Mean_Productivity=Corn_analysis_df.groupby(by='Year').agg({
    'Productivity (ku/ha)':'mean'
})
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(
    data=Mean_Productivity,
    x=Mean_Productivity.index,
    y='Productivity (ku/ha)',
    marker='o'
)
plt.title('Average Corn Producivity each Year')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```

### Insight
Based on the visualization of corn productivity per year in Indonesia, it can be concluded that corn productivity has increased every year. In the 2022  2023 period, there was the highest increase in productivity, which was 3% compared to other periods. Meanwhile, in the 2023  2024 period, corn productivity still increased, although relatively smaller.

## How Does the Average Corn Production Compare Between Provinces in Indonesia?

```python
Province_production=Corn_analysis_df.groupby(by='Province').agg({
  'Production (Ton)':'mean'
}).sort_values(by='Production (Ton)', ascending=False).head(10)
```

```python
sorted_data=Province_production.sort_values(by='Production (Ton)', ascending=True)

plt.figure(figsize=(8,6))
plt.barh(sorted_data.index, sorted_data['Production (Ton)'])
plt.title('Average Corn Production By Top 10 Province')
plt.xlabel('Average of Corn Production (Ton)')
plt.ylabel('Province')
plt.tight_layout()
plt.show()
```

### Insight
Based on the visualization of the 10 provinces with the highest corn production, it can be concluded that East Java Province is the largest corn producer in Indonesia. Meanwhile, among the 10 provinces with the highest corn production, South Sumatra Province has the lowest production compared to other provinces in the top 10 group.

## How Does Corn Production Compare Between Islands in Indonesia?

```python
Island_Corn=Corn_analysis_df.groupby(by='Island').agg({
    'Production (Ton)':'sum'
}).sort_values(by='Production (Ton)', ascending=False)
```

```python
labels=Island_Corn.index
sizes=Island_Corn['Production (Ton)']

percent=(sizes/sizes.sum()) *100
legend_labels=[f"{label} : {p:.1f}%" for label, p in zip(labels, percent)]
plt.figure(figsize=(6,6))
plt.pie(
    sizes,
    autopct=None,
    startangle=140,
    wedgeprops={'edgecolor':'black'}
)

plt.legend(
    legend_labels,
    title='Island',
    loc='center left',
    bbox_to_anchor=(1,0.5),
    fontsize=10
)
plt.title("Total Corn Production Distributions per Island", fontsize=14)
plt.tight_layout()
plt.show()
```

### Insight
Based on the visualization of rice production by island in Indonesia, it can be concluded that Java is the center of national rice production. This island contributes more than half of Indonesia's total rice production, namely 51.9%. In contrast, Papua and Maluku contribute very little to national rice production, only 0.1% each.

```python
Province_productivity=Corn_analysis_df.groupby(by='Province').agg({
    "Productivity (ku/ha)":"mean"
}).sort_values(by='Productivity (ku/ha)', ascending=False).head(10)
```

```python
sorted_Productivity=Province_productivity.sort_values(by='Productivity (ku/ha)', ascending=True)

plt.figure(figsize=(8,6))
plt.barh(sorted_Productivity.index, sorted_Productivity['Productivity (ku/ha)'])
plt.ylabel('Province')
plt.xlabel('Average of Corn Productivity')
plt.title('Average Corn Productivity Per Province')
plt.tight_layout()
plt.show()
```

### Insight
Based on the visualization of corn productivity by province, it can be concluded that East Java Province has the highest average corn productivity among the 10 provinces with the highest productivity in Indonesia. Meanwhile, West Java Province ranks last among the 10 provinces with the highest corn productivity.
