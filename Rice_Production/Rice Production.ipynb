# Preprocessing

```python
Rice_2018=Rice_2018.drop(index=[0,1,2,41])
Rice_2019=Rice_2019.drop(index=[0,1,2,41])
Rice_2020=Rice_2020.drop(index=[0,1,2,41])
Rice_2021=Rice_2021.drop(index=[0,1,2,41])
Rice_2022=Rice_2022.drop(index=[0,1,2,41])
Rice_2023=Rice_2023.drop(index=[0,1,2,41])
Rice_2024=Rice_2024.drop(index=[0,1,2,41])
```

```python
Rice_2018['Year']=2018
Rice_2019['Year']=2019
Rice_2020['Year']=2020
Rice_2021['Year']=2021
Rice_2022['Year']=2022
Rice_2023['Year']=2023
Rice_2024['Year']=2024
```

```python
Rice_Production_df=pd.concat([
    Rice_2018,
    Rice_2019,
    Rice_2020,
    Rice_2021,
    Rice_2022,
    Rice_2023,
    Rice_2024
])
```

```python
Rice_Production_df=Rice_Production_df.rename(columns={
    "38 Provinsi":'Province',
    "Unnamed: 1":'Harvest Area(ha)',
    "Unnamed: 2":"Productivity (ku/ha)",
    "Unnamed: 3":"Production(ton)"
})
```

```python
Rice_Production_df['Harvest Area(ha)']=Rice_Production_df['Harvest Area(ha)'].replace("-",0)
Rice_Production_df['Productivity (ku/ha)']=Rice_Production_df['Productivity (ku/ha)'].replace("-",0)
Rice_Production_df['Production(ton)']=Rice_Production_df['Production(ton)'].replace("-",0)
```

```python
Rice_Production_df['Harvest Area(ha)']=Rice_Production_df['Harvest Area(ha)'].astype(float)
Rice_Production_df['Productivity (ku/ha)']=Rice_Production_df['Productivity (ku/ha)'].astype(float)
Rice_Production_df['Production(ton)']=Rice_Production_df['Production(ton)'].astype(float)
```

# EDA

## Annual Rice Production in  Indonesia

```python
Rice_Production=Rice_Production_df.groupby(by='Year').agg({
    "Production(ton)":"sum"
})
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(
    data=Rice_Production,
    x=Rice_Production.index,
    y='Production(ton)',
    marker="o"
)
plt.title("Annual Rice Production in Indonesia")
plt.xlabel("Year")
plt.ylabel("Production (ton)")
plt.grid(True)
plt.show()
```

### Insight

Based on the visualization above, it can be concluded that rice production has been declining from 2018 to 2024, with a total decrease of around 10,23% or around 6,05 million tons . The most significant decline occurred between 2018 and 2019, which amounts to 4,6 million tons





## Average of Harvest Area in Indonesia Per Year

```python
Harvest_area=Rice_Production_df.groupby(by='Year').agg({
    "Harvest Area(ha)":"mean"
})
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(
    data=Harvest_area,
    x=Harvest_area.index,
    y='Harvest Area(ha)',
    marker="o"
)
plt.title('Average Rice Field Area in Indonesia')
plt.xlabel('Year')
plt.ylabel('Rice Field (ha)')
plt.show()
```

## Insight:

Based on the visualization of rice field area in Indonesia from 2018 to 2024, it can be seen that the area has decreased every year. In 2018, the rice field area reached around 300,000 hectares. However, in the following years, the area continued to decline gradually until it reached its lowest point in 2024. The decreasing amount of rice fields from 2018 to 2024 is around 11.7% or around 35000 hectares



## Average Rice Production Productivity in Indonesia Per Year

```python
Production=Rice_Production_df.groupby(by='Year').agg({
    "Productivity (ku/ha)":"mean"
})
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=Production,
             x=Production.index,
             y='Productivity (ku/ha)',
             marker="o")
plt.title('Average Rice Productivuty (ku/ha) in Indonesia')
plt.xlabel('Year')
plt.ylabel('Productivity(ku/ha)')
plt.show()
```

### Insight
Based on the visualization above, productivity fluctuated between 2018 and 2024. In the period from 2022 to 2024, there was siginificant increase in productivity of around 4.22 ku/ha. Meanwhile, between 2018 and 2020, there was a decline in productivity of 1.26 ku/ha

## Top Rice Producing Provinces in Indonesia

```python
Rice_Province=Rice_Production_df.groupby(by='Province').agg({
    "Production(ton)":"mean"
}).sort_values(by="Production(ton)", ascending=False).head(10)
```

```python
sorted_data=Rice_Province.sort_values(by='Production(ton)', ascending=True)

plt.figure(figsize=(10,8))
plt.barh(sorted_data.index, sorted_data['Production(ton)'])
plt.ylabel('Province')
plt.xlabel('Avarage of Rice Production(Ton)')
plt.title('Average Rice Production per Province')
plt.tight_layout()
plt.show()
```

### Insight
Based on the visualization of the 10 rice-producing provinces in Indonesia, there are three provinces that most rice, which include East Java, Central Java and West Java, which dominate rice production on the island of java

```python
Island_Rice=Rice_Production_df.groupby(by='Island').agg({
    "Production(ton)":"sum"
})
```

## Distribution of Rice Production by Island in Indonesia

```python
labels=Island_Rice.index
sizes=Island_Rice['Production(ton)']

percent=(sizes/sizes.sum()) *100
legend_labels=[f"{label}: {p:.1f}%" for label, p in zip(labels, percent)]

plt.figure(figsize=(8,8))
plt.pie(
    sizes,
    autopct=None,
    startangle=140,
    wedgeprops={'edgecolor':'black'}
)

plt.legend(
    legend_labels,
    title='Island',
    loc='center left',
    bbox_to_anchor=(1,0.5),
    fontsize=10
)
plt.title("Total Rice Production Distributions per Island", fontsize=14)
plt.tight_layout()
plt.show()
```

### Insight
Based on the visualization of Rice Production Distributions Per Islands, Java Island dominates more than half of Indonesia's rice production, contributing around 55.6% of the national total. In addition, Sumatra island contributes around 20.7% and Sulawesi Island contributes around 13.7%

### Conclusion
According to analysis, it can be concluded that rice production has been decreasing year by year, along with the decline in rice field area. This shows that rice production and rice field area have a relationship: when the area decreases, production also decreases. Meanwhile, rice productivity in Indonesia has fluctuated during the period from 2018 to 2024. In addiction, the three largest rice-producting provinces in Indonesia are East Java, Central Java and West Java, all of which are located on the island of Java.
